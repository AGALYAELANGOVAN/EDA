```python
df=df.drop_duplicates()
df                          #to emove duplicates
```

```python
df.drop(df[(df['gender'] == "Other")].index)
df
```

```python
df["gender"].unique()
```

```python
df["smoking_history"].unique()
```

```python
df['smoking_history'] = df['smoking_history'].replace('not current', 'former')
df
```

**LABEL ENCODING**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
gender=LabelEncoder()
smoker=LabelEncoder()
df["gender"]=gender.fit_transform(df.gender)
df["smoking_history"]=smoker.fit_transform(df.smoking_history)
df.head(5)
```

```python
x=df[["gender","age","hypertension","heart_disease","smoking_history","bmi","HbA1c_level","blood_glucose_level"]]
y=df[["diabetes"]]
x.head(5),y.head(5)
```

**SPILITING INTO TRAINING AND TESTING**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.30, random_state=70)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

**TRAINING THE MODEL USING LOGISTIC REGRESSION**

```python
def sigmoid(x):                  #ccreating logistic function
    return 1/(1+np.exp(-x))
```

```python
sigmoid(-100000000000)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_logit=LogisticRegression(max_iter=500)
```

```python
model_logit.fit(X_train,y_train)
```

**TESTING THE DATA**

```python
y_pred=model_logit.predict(X_test)
```

```python
model_logit.coef_*100
```

**FOR FINDING ACCURACY SCORE**

```python
from sklearn.metrics import accuracy_score

accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_test,y_pred)
conf_mat
```
