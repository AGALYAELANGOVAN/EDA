```python
df.drop("RowNumber", axis =1, inplace=True) # Here we are removing the Row Number column alone as it is not a suitable data for our EDA, it as the row id number only
```

```python
df
```

```python
df['Encoded_Surname'] = df['Surname'].astype('category').cat.codes
df['Encoded_Geography'] = df['Geography'].astype('category').cat.codes
df['Encoded_Gender'] = df['Gender'].astype('category').cat.codes
df['Encoded_Card Type'] = df['Card Type'].astype('category').cat.codes
```

```python
df
```

```python
updated_df = df.drop(["Surname", "Geography", "Gender", "Card Type"], axis =1) # For creating the Correlation and Heatmap, we need only the integer so
# for that we are removing the string column data
```

```python
updated_df
```

```python
correlation_data = updated_df.corr().round(2)
correlation_data
```

```python
plt.figure(figsize=(18,8))
sns.heatmap(correlation_data.corr().round(2), annot = True, cmap = 'Blues', cbar = True, vmin = -1, vmax = 1)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
min_val = 0
max_val = 1
filtered_correlation_data = correlation_data.where((correlation_data >= min_val) & (correlation_data <= max_val))
```

```python
plt.figure(figsize=(18,8))
sns.heatmap(filtered_correlation_data, annot = True, cmap = 'Blues', cbar = True, vmin = -1, vmax = 1)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

# EDA

```python
card_type = df['Geography'].value_counts()
value_count = card_type.values
unique_words = card_type.index
explode = (0.1,0,0)
plt.figure(figsize=(4, 4))
plt.pie(x=value_count, labels=unique_words, autopct='%1.1f%%', explode = explode)
plt.title('Distribution of Geography')
plt.show()
```

```python
satisfaction_counts = df["Exited"].value_counts().sort_values(ascending=False)
satisfaction_counts.plot(kind="bar")
plt.title("Bar Graph to show the Customer's Who Exited and Current Customer")
plt.xlabel("Satisfaction Score")
plt.ylabel("Count of Satisfaction Score")
plt.show()
```

# Taking and Analzing the Highest Geography Bank Chunk Data

```python
france_data = df[df['Geography'] == 'France']
france_data = france_data.drop(['Encoded_Surname', 'Encoded_Geography', 'Encoded_Gender', 'Encoded_Card Type'], axis=1)
```

```python
card_type = france_data['Gender'].value_counts()
value_count = card_type.values
unique_words = card_type.index
plt.figure(figsize=(4, 4))
plt.pie(x=value_count, labels=unique_words, autopct='%1.1f%%')
plt.title('Distribution of Gender')
plt.show()
```

```python
card_type = france_data['Card Type'].value_counts()
value_count = card_type.values
unique_words = card_type.index
plt.figure(figsize=(4, 4))
plt.pie(x=value_count, labels=unique_words, autopct='%1.1f%%')
plt.title('Distribution of Card Types')
plt.show()
```

```python
france_male_data = france_data[france_data['Gender'] == 'Male']
france_male_data
```

```python
sns.displot(france_male_data.Age, color = "blue")
plt.title("Distribution Plot for Male Customer Based on Age")
plt.show()
```

```python
pvt_count = pd.pivot_table(france_male_data, index=["Card Type"], values=["Point Earned"], aggfunc="sum").sort_values(by="Point Earned", ascending = False)
```

```python
plt.plot(pvt_count.index, pvt_count.values, color = 'c', marker = '*')
plt.title("The Total Points Earned Across Each Card Type")
plt.xlabel("Card Type")
plt.ylabel("Total Points Earned")
plt.show()
```

```python
pvt_satisfaction_score = france_male_data.groupby(['Card Type', 'Satisfaction Score'])['Satisfaction Score'].sum()
```

```python
pvt_satisfaction_score.plot(kind='bar', stacked=True, figsize=(10, 7))
plt.title('Sum of Satisfaction Scores by Card Type and Score')
plt.xlabel('Card Type')
plt.ylabel('Sum of Satisfaction Scores')
plt.legend(title="Satisfaction Score")
plt.show()
```
