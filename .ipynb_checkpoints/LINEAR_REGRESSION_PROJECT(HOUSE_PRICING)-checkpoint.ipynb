```python
X=data2.drop(["SALES_PRICE"],axis=1)
Y=data2[["SALES_PRICE"]]
X_train,X_test,Y_train,Y_test=tts(X,Y,test_size=0.30,random_state=70)
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
model_lr=LR()
model_lr
model_lr.fit(X_train,Y_train)
```

```python
X_te=data2.drop(["SALES_PRICE"],axis=1)
Y_te=data2[["SALES_PRICE"]]
X_te,Y_te
```

```python
Y_pred=model_lr.predict(X_te)
Y_pred
```

```python
r2_score=r2(Y_te,Y_pred) 
r2_score*100
```

```python
mse1= mse(Y_te, Y_pred)   #error value
mse1*100
```

```python
accuracy = 1-mse1   #accuracy value
accuracy*100
```
